```python
import os
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Timeline
from pyecharts.charts import Bar3D
```

```python
src_file = r'../../data/pre2011_data.csv'
if os.path.exists(src_file):
    data = pd.read_csv(src_file, header=None, encoding='latin-1')
else:
    raise FileNotFoundError("File %s not found!" % (src_file))
data.dropna(axis=1, how='all', inplace=True)
data.fillna(0, inplace=True)

num_cols = len(data.columns)
num_days = 9
# num_points_perday = 172800
num_points_perday = 512
```

```python
color0 = ['#FF76A2']

f_1 = Timeline()
# total_lenth = num_points_perday
total_lenth = 1

bar_data = []
for t in range(total_lenth):
    row_data = data.loc[t]
    slice_data = []
    for i in range(num_cols):
        tmp_y = i // 7
        tmp_x = i % 7
        y_data = tmp_y
        x_data = tmp_x
        z_data = row_data.iloc[i]
        slice_data.append([x_data, y_data, z_data])
    bar_data.append(slice_data)

cable_id2name_y = ["SJS", "SJX"]
cable_id2name_x = ["08", "09", "10", "11", "12", "13", "14"]

for i in range(0, total_lenth):
    # x_max = i
    # y_max = int(max(col_data[0:i+1])) + 10
    # y_min = int(min(col_data[0:i+1])) -10
    z_max = max(list(item[2] for item in bar_data[i]))

    c = (
        Bar3D()
        .add(
            series_name="full field",
            data=bar_data[i],
            xaxis3d_opts=opts.Axis3DOpts(type_="category", data=cable_id2name_x),
            yaxis3d_opts=opts.Axis3DOpts(type_="category", data=cable_id2name_y),
            zaxis3d_opts=opts.Axis3DOpts(type_="value"),
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Full Field Diagram"),
            visualmap_opts=opts.VisualMapOpts(
                max_=z_max,
                range_color=[
                    "#313695",
                    "#4575b4",
                    "#74add1",
                    "#abd9e9",
                    "#e0f3f8",
                    "#ffffbf",
                    "#fee090",
                    "#fdae61",
                    "#f46d43",
                    "#d73027",
                    "#a50026",
                ],
            )
        )
        .render("full_field_cfr_tsa.html")
        # .render_notebook()
    )
    # f_1.add(c, "{}".format(i))
    # f_1.add_schema(axis_type='value', play_interval=1000,
    #                pos_bottom="-5px", pos_left="left", pos_right="5%",
    #                label_opts=opts.LabelOpts(is_show=False),
    #                is_loop_play=False)
    # print("frame:{:>3d} is done".format(i))
# f_1.render("full_field_cfr_tsa.html")
# f_1.render_notebook()
```
