```python
import os
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Timeline
from pyecharts.charts import HeatMap
```

```python
src_file = r'../../data/pre2011_data.csv'
if os.path.exists(src_file):
    data = pd.read_csv(src_file, header=None, encoding='latin-1')
else:
    raise FileNotFoundError("File %s not found!" % (src_file))
data.dropna(axis=1, how='all', inplace=True)
data.fillna(0, inplace=True)

num_cols = len(data.columns)
num_days = 9
num_points_perday = 172800
# num_points_perday = 512
```

```python
color0 = ['#FF76A2']

f_1 = Timeline()
# total_lenth = num_points_perday
total_lenth = 1920

bar_data = []
for t in range(total_lenth):
    row_data = data.loc[t]
    slice_data = []
    for i in range(num_cols):
        tmp_y = i // 7
        tmp_x = i % 7
        y_data = tmp_y
        x_data = tmp_x
        z_data = row_data.iloc[i]
        slice_data.append([x_data, y_data, z_data])
    bar_data.append(slice_data)

cable_id2name_y = ["SJS", "SJX"]
cable_id2name_x = ["08", "09", "10", "11", "12", "13", "14"]
index_y = list(range(2))
index_x = list(range(7))

for i in range(0, total_lenth):

    # z_max = max(list(item[2] for item in bar_data[i])) + 100
    # z_min = min(list(item[2] for item in bar_data[i])) - 100
    z_max = 3200
    z_min = 2000

    c = (
        HeatMap()
        .add_xaxis(xaxis_data=index_x)
        .add_yaxis(series_name="Full Field Diagram", yaxis_data=index_y, value=bar_data[i],
                   label_opts=opts.LabelOpts(position="middle"))

        .set_global_opts(
            title_opts=opts.TitleOpts(title=""),
            visualmap_opts=opts.VisualMapOpts(
                min_=z_min, max_=z_max,
                pos_bottom="center", pos_right="right",
                range_color=[
                    "#313695",
                    "#4575b4",
                    "#74add1",
                    "#abd9e9",
                    "#e0f3f8",
                    "#ffffbf",
                    "#fee090",
                    "#fdae61",
                    "#f46d43",
                    "#d73027",
                    "#a50026",
                ],
            ),
        )
    )
    f_1.add(c, "{}".format(i))
    f_1.add_schema(axis_type='value', play_interval=30,
                   pos_bottom="-5px", pos_left="left", pos_right="5%",
                   label_opts=opts.LabelOpts(is_show=False),
                   is_loop_play=False)
    
f_1.render("full_field_cfr_tsa.html")
# f_1.render_notebook()
```
